TRAINING DATA FOR MODEL

```python
X_train=training[:,1:]/255
y_train=training[:,0]
```

```python
X_test=testing[:,1:]/255
y_test=testing[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_validate,y_train,y_validate=train_test_split(X_train,y_train,test_size=0.2,random_state=12345)
```

```python
X_train.shape
```

```python
X_validate.shape
```

```python
X_train
```

```python
X_train=X_train.reshape(X_train.shape[0],*(28,28,1))
X_test=X_test.reshape(X_test.shape[0],*(28,28,1))
X_validate=X_validate.reshape(X_validate.shape[0],*(28,28,1))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_validate.shape
```

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Conv2D,Conv3D,MaxPool2D,Flatten,Dense,Dropout
from keras.optimizers import Adam
```

```python
our_model=Sequential()
our_model
```

```python
our_model.add(Conv2D(32,3,3, input_shape=(28,28,1),activation='relu'))
```

```python
our_model.add(MaxPool2D(pool_size=(2,2)))
```

```python
our_model.add(Flatten())
```

```python
our_model.add(Dense(units=32,activation='relu'))
```

```python
our_model.add(Dense(units=10,activation='sigmoid'))
```

```python
#importing sparse_categorical_crossentropy from tensorflow
tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=False,
    name='sparse_categorical_crossentropy'
)
our_model.compile(loss='sparse_categorical_crossentropy',optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
epochs=50
```

```python
our_model.fit(X_train,
    y_train,
    batch_size=None,
    epochs=epochs,
    verbose=1,
    validation_data=(X_validate,y_validate))
```

```python
evaluate_m=our_model.evaluate(X_test,y_test)
```

```python
print(f'Test Accuracy {round(evaluate_m[1],2)}')
```

```python
predict_x=our_model.predict(X_test) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
classes_x
```

```python
W_grid = 8
L_grid = 8

fig, axes = plt.subplots(L_grid, W_grid, figsize = (16,16))
axes = axes.ravel() # flaten the 12 x 12 matrix into 225 array
# Select a random number from 0 to n_train
for i in np.arange(0, W_grid * L_grid): # create evenly spaces variables 
    # read and display an image with the selected index    
    axes[i].imshow(X_test[i].reshape((28,28)) )
    axes[i].set_title('Prediction={:0.1f}\nTrue={:0.1f}'.format(classes_x[i],y_test[i]), fontsize = 9)
    axes[i].axis('off')

plt.subplots_adjust(hspace=0.4)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn=confusion_matrix(y_test,classes_x)
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(cn,annot=True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,classes_x))
```
